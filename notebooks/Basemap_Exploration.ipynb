```python
#https://stackoverflow.com/questions/18109859/how-to-install-matplotlib-basemap-module-on-windows-7-with-winpython-or-any-pyt
import matplotlib.pyplot as plt
import matplotlib.cm
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
```

```python
m = Basemap(resolution = "c", #c = crude, l = low, i = intermediate, h = high, f = full or None
           projection = "merc",
           lat_0 = 64.789785, lon_0 = -150.996703,
           llcrnrlon = -169.49, llcrnrlat = 53.4, urcrnrlon = -129.4, urcrnrlat = 71.63)
```

```python
m.drawmapboundary()
m.fillcontinents()
m.drawcoastlines()
```
