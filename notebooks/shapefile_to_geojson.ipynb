```python
def record_to_json(shape_record,fields):
    attributes = dict(zip(fields,shape_record.record))
    geometry = shape_record.shape.__geo_interface__
    return dict(type='Feature',geometry=geometry,properties=attributes)
```

```python
fields = [x[0] for x in sf.fields[1:]]
geojson = [record_to_json(shape_record, fields) for shape_record in shape_records]
```

```python
len(geojson)
```

```python
import json
writer = open("..\\tep\\data\\geospatial\\Japan\\MBI_JP_R2017_Shikuchoson.json","w")
writer.write(json.dumps({"type": "FeatureCollection","features": geojson}))
writer.close()
```
